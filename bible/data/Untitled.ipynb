```python
import re
with open("new_king_james_bible.html", 'a') as f:
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
        pattern = r"\d+:\d+"
        
        # Find all occurrences using finditer
        all_matches = re.finditer(pattern, text)
        
        # Extract locations and store them in a list
        locations = []
        for match in all_matches:
          # match.start() gives the starting index of the match
          locations.append(match.start())
        
        
        for i, l in enumerate(locations):
            if i == 0:
                text_new = text[:locations[i+1]].replace("<p>", '').replace("</p>" ,"").strip()
            if i == len(locations) - 1:
                text_new = text[locations[i]:-1].replace("<p>", '').replace("</p>" ,"").strip()
            else:
                text_new = text[locations[i]:locations[i+1]].replace("<p>", '').replace("</p>" ,"").strip()

            if "</div>" in text_new:
                split = re.split("</div>", text_new)
                text_new = f"<p>{split[0].strip()}</p></div>\n\n{split[1]}" 
                print(text_new)
            else:
                text_new = f"<p>{text_new}</p>"
                    
                
            # Write to file
            f.write(text_new)
            
print("Complete")
```
